```python
import os
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnableSequence
from langchain.llms import HuggingFaceEndpoint

# Set Hugging Face API token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "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"

# Define a summarization-specific prompt template
prompt = PromptTemplate.from_template(
    "Summarize the following text:\n\n{context}"
)

# Initialize Hugging Face Endpoint (facebook/bart-large-cnn is a summarization model)
hub_llm = HuggingFaceEndpoint(
    endpoint_url="https://router.huggingface.co/hf-inference/models/facebook/bart-large-cnn",
    task="summarization"
)

# Construct the LangChain RunnableSequence
pipeline = RunnableSequence(first=prompt, last=hub_llm)

# Example context for summarization
context = """
The New Orleans Saints defeated the Indianapolis Colts 31-17 in Super Bowl XLIV,
held on February 7, 2010, at Sun Life Stadium in Miami Gardens, Florida.
This marked the first Super Bowl victory for the Saints in the franchise's history.
"""

try:
    # Run the pipeline with a summarization-specific input
    response = pipeline.invoke({"context": context})
    print("Generated Summary:", response)
except Exception as e:
    print(f"An error occurred: {e}")
```
